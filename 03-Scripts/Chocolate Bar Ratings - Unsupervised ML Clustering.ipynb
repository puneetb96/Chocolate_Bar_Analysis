```python
#Create subset dataframe with only necessary numerical columns
df_numerical = df[['Cocoa_Percentage','Rating']]

#Show subset dataframe
df_numerical.head()
```

```python
#Using StandardScaler library to standardize numerical dataframe, for ML analysis
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_numerical)

#Show scaled values
print('Scaled Values: \n', df_scaled)
```

- Data prepped for elbow technique

## 3. The Elbow Technique

```python
#Define the range of potential clusters in the data
num_cl = range(1,10)

#Defines k-means clusters from range above
kmeans = [KMeans(n_clusters = i) for i in num_cl]
```

```python
# Create a score that represents a rate of variation for the given cluster option
score = [kmeans[i].fit(df_scaled).score(df_scaled) for i in range(len(kmeans))]

#Show score
score
```

```python
#Plot elbow curve using Pylab
pl.plot(num_cl,score, marker='o')

#Titles and Labels
pl.xlabel('Number of Clusters', fontsize=12)
pl.ylabel('Score', fontsize=12)
pl.title('Elbow Curve', fontsize=14)


#Save Image
pl.savefig(os.path.join(path, '04-Analysis','Data Visualizations','elbow_curve_plot.png'),
                        bbox_inches='tight',
                        dpi=300
          )

#Show plot
pl.show()
```

- The Elbow curve shows a dramatic increase of up to 4 clusters, before flattening out
- There is a small jump between 5 and 6 clusters, but we'll ignore this for now
#### *The number of clusters we will choose for further analysis will be 4*

## 4. K-means clustering

```python
#Create the k-means object with 4 clusters
kmeans = KMeans(n_clusters = 4)
```

```python
#Fit the k-means object to the data
kmeans.fit(df_scaled)
```

```python
#Add 'Clusters' column to numerical dataframe
df_numerical['Clusters'] = kmeans.fit_predict(df_numerical)
```

```python
#Check clusters in dataframe head
df_numerical.head()
```

```python
#Check count of 'Clusters' column values
df_numerical['Clusters'].value_counts()
```

```python
#Plot the clusters for 'Ratings' and 'Cocoa_Percentage' variables
plt.figure(figsize=(12,8))
ax = sns.scatterplot(                        #Scatterplot
    x = df_numerical['Cocoa_Percentage'],    #Subsetting x and y arguments to avoid using their labels
    y = df_numerical['Rating'],
    hue = kmeans.labels_,                     #Use kmeans.label for hue to show clusters, result of running the k-means algorithm
    s = 100)                                 # 'S' shows size of the points we see inthe plot


#Titles, labels, and grid lines
plt.grid(False)                              #Removes gridlines from the background
plt.title('Clusters for Cocoa Percentages vs. Ratings', fontsize=14)
plt.xlabel('Cocoa Percentage', fontsize=12)
plt.ylabel('Rating', fontsize=12)

#Save plot
plt.savefig(os.path.join(path, '04-Analysis', 'Data Visualizations', 'cluster_chart.png'),
            dpi = 300,
            bbox_inches = 'tight'
           )
#Show Plot
plt.show()
```

```python
#Define color for each cluster
df_numerical.loc[df_numerical['Clusters'] == 0, 'Cluster'] = 'Pink'
df_numerical.loc[df_numerical['Clusters'] == 1, 'Cluster'] = 'Light Purple'
df_numerical.loc[df_numerical['Clusters'] == 2, 'Cluster'] = 'Dark Purple'
df_numerical.loc[df_numerical['Clusters'] == 3, 'Cluster'] = 'Black'
```

```python
#Check counts for each cluster color
df_numerical['Cluster'].value_counts()
```

```python
#Grouby and find descriptive statistics for each cluster
df_numerical.groupby('Cluster').agg({'Cocoa_Percentage': ['mean','median','max','count'],
                                     'Rating': ['mean', 'median','max','count']
                                    })
```

### Analysis:

#### Pink Cluster (0):
- This cluster represents cocoa percentages that fall between 65-75%.
- It has a median and mean of 70% cocoa percentage (naturally), and has the highest mean rating of 3.25
- This portion also represents the majority of chocolate bars rated, at over 1020. This seems to be the percentage range of chocolates that are produced the most, and best tasting.
- It is the only percentage that has a rating of 5.0 for a chocolate bar (Elite), and contains the most 4.0 ratings
- This cluster could be described as the "best in class" percentage for chocolate tasting

#### Light Purple Cluster (1):
- This cluster represents cocoa percentages 85% and above
- There are few bars produced with this percentage of chocolate (only 89 bars), which means chocolate makers don't typically produce such high cocoa wuality chocolate
- The mean and median scores for this percentage are the lowest of the 4 clusters, and is the only cluster with a mean rating score below 3.0 (2.79)
- The "worst" tasting chocolate could be categorized in this cluster, and chocolate makers should avoid high percentage of cocoa

#### Dark Purple Cluster (2):
- This cluster represents cocoa percentages 65% and below (as low as 42%).
- There are about 244 chocolate bars that fall under this category, so a modest portion of the total clusters
- While this cluster represents the other extreme of the spectrum of cocoa percentage, it performs better on a rating scale overall than the light purple cluster (high cocoa percentage), with a mean score of 3.13
- While not as successful as the pink cluster, the dark purple cluster of chococlates produced are still popular amongst tasters

#### Black Cluster (3):
- This cluster represents cocoa percentages between 75-85%
- There 442 chocolate bars that fall under this category, which is the second largest cluster behind pink (0)
- While larger than the dark purple cluster, it is the most similar in mean ratings (taste) to the dark purple cluster with a 3.13 mean score.
- Both black and dark purple clusters are good tasting chocolates, but both fall in second place to a clear winner of the pink cluster

#### *Overall, the best chocolate produced falls in the 65-75% cocoa percentage range. This should be the primary focus of all chocolate makers. They can also produce chocolate that can fall as low as 65%, for moderate results as well. It is highly advised not to produce chocolate that has higher than an 85% cocoa percentage because the unique bitter taste might not be for everyone (even the tasting experts).*
